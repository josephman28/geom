#### a)
Show that $\pi : S^2\setminus\{N\} \rightarrow \mathbb{R}^2$ is given by $X = \frac{x}{1 - z}$, $Y = \frac{y}{1 - z}$.

  The line that passes through the north pole $\{N\} = (0, 0, 1)$ and point $(X, Y, 0)$ is given by the parametric equations:

  $x_1 = tX, \quad y_1 = tY, \quad z_1 = 1 - t, \quad -\infty < t < \infty$
  
  Note that the line cuts the sphere precisely when $t$ satisfies:
  
  $x_1^2 + y_1^2 + z_1^2 = t^2 X^2 + t^2 Y^2 + (1-t)^2 = t^2(X^2+Y^2+1) + 1 - 2t = 1$
  
  The roots of this equation are $t = 0$ (the north pole) and 
  
  $t = \frac{2}{X^2 + Y^2 + 1}$.

  Using that parametrization, it is possible to represent the stereographic projection $(X, Y)$ in the plane $xy$ as:

  $X = \frac{x_1}{t}, \quad Y = \frac{y_1}{t}, \quad t = 1 - z_1$
  
  Substituting $t$, we have:
  
  $X = \frac{x_1}{1 - z_1}, \quad Y = \frac{y_1}{1 - z_1}$
  


#### b)
Show that it is possible, using stereographic projection, to cover the sphere with two coordinate neighborhoods.

  As the stereographic projection $\pi$ is a bijective function, its inverse exists, that is $\pi^{-1}$, defined by:
 
  $x_1 = tX, \quad y_1 = tY, \quad z_1 = 1 - t, \quad -\infty < t < \infty$
  
  Simplifying these equations, we have:

  $x_1 = \frac{2X}{1 + X^2 + Y^2}, \quad y_1 = \frac{2Y}{1 + X^2 + Y^2}, \quad z_1 = \frac{-1+ X^2 - Y^2}{1 + X^2 + Y^2}.$
  
 
  This function maps the entire plane $xy$ to the entire sphere $S^2$ except for the center of projection, which corresponds to the north pole. To totally cover the entire sphere $S^2$, we can define $\alpha : \mathbb{R}^2 \rightarrow S^2\setminus\{S\}$ similarly to $\pi^{-1}$ but with the line passing through $\{S\} = (0, 0, -1)$ instead of $(0, 0, 1)$. $\alpha$ is given by:
  
  $x_2 = tX, \quad y_2 = tY, \quad z_2 = t - 1, \quad -\infty < t < \infty$

  Simplifying these equations, we have:
  
  $x_2 = \frac{2X}{1 + X^2 + Y^2}, \quad y_2 = \frac{2Y}{1 + X^2 + Y^2}, \quad z_2 = \frac{1 - X^2 - Y^2}{1 + X^2 + Y^2}$
  
  In this way, $\pi^{-1} \cup \alpha$ covers the sphere with two coordinate neighborhoods.

To evaluate the first fundamental form for the ellipsoid parametrization, we define the parametrization:


\begin{aligned}
x &= a \sin \theta \cos \phi \\
y &= b \sin \theta \sin \phi \\
z &= c \cos \theta
\end{aligned}


Then, we calculate the tangent vectors:


\begin{aligned}
\vec{x}_\theta &= \frac{\partial \vec{x}}{\partial \theta} = \left(\frac{\partial x}{\partial \theta}, \frac{\partial y}{\partial \theta}, \frac{\partial z}{\partial \theta}\right) \\
\vec{x}_\phi &= \frac{\partial \vec{x}}{\partial \phi} = \left(\frac{\partial x}{\partial \phi}, \frac{\partial y}{\partial \phi}, \frac{\partial z}{\partial \phi}\right)
\end{aligned}


After differentiating each component of ($\vec{x}(\theta, \phi)$) with respect to ($\theta$) and ($\phi$), we find:


\begin{aligned}
\vec{x}_\theta &= (a \cos \theta \cos \phi, b \cos \theta \sin \phi, -c \sin \theta) \\
\vec{x}_\phi &= (-a \sin \theta \sin \phi, b \sin \theta \cos \phi, 0)
\end{aligned}


Next, we calculate the dot products of these tangent vectors:


\begin{aligned}
\vec{x}_\theta \cdot \vec{x}_\theta &= a^2 \cos^2 \theta \cos^2 \phi + b^2 \cos^2 \theta \sin^2 \phi + c^2 \sin^2 \theta \\
\vec{x}_\theta \cdot \vec{x}_\phi &= -a^2 \sin \phi \sin \theta \cos \phi \cos \theta + b^2 \sin\phi \sin \theta \cos \phi \cos \theta \\
\vec{x}_\phi \cdot \vec{x}_\phi &= a^2 \sin^2 \theta \sin^2 \phi + b^2 \sin^2 \theta \cos^2 \phi
\end{aligned}

\begin{aligned}
E &= \vec{x}_\theta \cdot \vec{x}_\theta \\
F &= \vec{x}_\theta \cdot \vec{x}_\phi \\
G &= \vec{x}_\phi \cdot \vec{x}_\phi
\end{aligned}

Therefore, the first fundamental form theorically is:

\begin{aligned}
E &= a^2 \cos^2 \theta \cos^2 \phi + b^2 \cos^2 \theta \sin^2 \phi + c^2 \sin^2 \theta \\
F &= -a^2 \sin \phi \sin \theta \cos \phi \cos \theta + b^2 \sin\phi \sin \theta \cos \phi \cos \theta \\
G &= a^2 \sin^2 \theta \sin^2 \phi + b^2 \sin^2 \theta \cos^2 \phi

Now let's see if it's the same that we obtain on the following Python Code:



```python
import sympy as sp

# Define the symbols
a, b, c, theta, phi = sp.symbols('a b c theta phi')

# Define the parametrization
x = a * sp.sin(theta) * sp.cos(phi)
y = b * sp.sin(theta) * sp.sin(phi)
z = c * sp.cos(theta)

# Calculate the tangent vectors
xt = sp.diff(x, theta)
yt = sp.diff(y, theta)
zt = sp.diff(z, theta)

xp = sp.diff(x, phi)
yp = sp.diff(y, phi)
zp = sp.diff(z, phi)

# Calculate the dot products of the tangent vectors
E = xt**2 + yt**2 + zt**2
F = xt * xp + yt * yp + zt * zp
G = xp**2 + yp**2 + zp**2

# Print the results
print("E =", E)
print("F =", F)
print("G =", G)
```

To evaluate the second fundamental form for the ellipsoid, let's denote the position vector as $\vec{X}(u, v)$ and the parameters as $u$ and $v$.

The position vector for the ellipsoid is given by:


$\vec{X}(u, v) = (a \sin u \cos v, b \sin u \sin v, c \cos u)$


where $a$, $b$, and $c$ are constants that determine the shape of the ellipsoid.

The second fundamental form is given by:


$II = e \, du^2 + 2f \, du \, dv + g \, dv^2$


where $e$, $f$, and $g$ are coefficients that depend on the second partial derivatives of $\vec{X}$.

To compute these coefficients, we need to evaluate the second partial derivatives of $\vec{X}$ with respect to $u$ and $v$. Let's calculate them:


\begin{aligned}
\frac{\partial^2 \vec{X}}{\partial u^2} &= (-a \sin u \cos v, -a \sin u \sin v, -c \cos u) \\
\frac{\partial^2 \vec{X}}{\partial u \partial v} &= (-a \cos u \sin v, a \cos u \cos v, 0) \\
\frac{\partial^2 \vec{X}}{\partial v^2} &= (-a \sin u \cos v, -a \sin u \sin v, 0)
\end{aligned}


Now, let's compute the coefficients:


\begin{aligned}
e &= \frac{\partial^2 \vec{X}}{\partial u^2} \cdot \vec{n} = (-a \sin u \cos v, -a \sin u \sin v, -c \cos u) \cdot \vec{n} \\
f &= \frac{\partial^2 \vec{X}}{\partial u \partial v} \cdot \vec{n} = (-a \cos u \sin v, a \cos u \cos v, 0) \cdot \vec{n} \\
g &= \frac{\partial^2 \vec{X}}{\partial v^2} \cdot \vec{n} = (-a \sin u \cos v, -a \sin u \sin v, 0) \cdot \vec{n}
\end{aligned}

where $\vec{n}$ is the unit normal vector to the ellipsoid surface.

Simplifying the expressions, we obtain:


\begin{aligned}
e &= -a^2 \sin u \cos u \cos v \cdot \frac{1}{\sqrt{a^2 \sin^2 u \cos^2 v + b^2 \sin^2 u \sin^2 v + c^2 \cos^2 u}} \\
f &= a^2 \sin u \cos u \sin v \cdot \frac{1}{\sqrt{a^2 \sin^2 u \cos^2 v + b^2 \sin^2 u \sin^2 v + c^2 \cos^2 u}} \\
g &= -a^2 \sin u \cos u \cos v \cdot \frac{1}{\sqrt{a^2 \sin^2 u \cos^2 v + b^2 \sin^2 u \sin^2 v + c^2 \cos^2 u}}
\end{aligned}


Therefore, the second fundamental form for the ellipsoid is:


$II = -a^2 \sin u \cos u \cos v \, du^2 + 2a^2 \sin u \cos u \sin v \, du \, dv - a^2 \sin u \cos u \cos v \, dv^2$

Consider a point $q \in U$. We want to evaluate the normal vector on the surface at $p = \vec{X}(q) \in \vec{X}(U)$, given by $\hat{N}(p) = \vec{X}_\phi \times \vec{X}_\theta / |\vec{X}_\phi \times \vec{X}_\theta|(p)$.

To calculate the normal vector, we need to find the partial derivatives $\vec{X}_\phi$ and $\vec{X}_\theta$, and then compute their cross product.

The partial derivatives of $\vec{X}$ are:


$\vec{X}_\phi = (a \cos \phi \sin \theta, b \sin \phi \sin \theta, 0)$


$\vec{X}_\theta = (-a \sin \phi \cos \theta, b \cos \phi \cos \theta, -c \sin \theta)$


Now, let's compute their cross product:


$\vec{X}_\phi \times \vec{X}_\theta = \begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
a \cos \phi \sin \theta & b \sin \phi \sin \theta & 0 \\
-a \sin \phi \cos \theta & b \cos \phi \cos \theta & -c \sin \theta
\end{vmatrix}$


Expanding the determinant, we get:


$\vec{X}_\phi \times \vec{X}_\theta = (bc \cos \phi \sin \theta, -ac \sin \phi \cos \theta, -ab \sin^2 \theta \cos \phi - ab \sin \theta \cos^2 \phi)$

Next, we need to calculate the magnitude of the cross product:


$|\vec{X}_\phi \times \vec{X}_\theta|(p) = \sqrt{(bc \cos \phi \sin \theta)^2 + (-ac \sin \phi \cos \theta)^2 + (-ab \sin^2 \theta \cos \phi - ab \sin \theta \cos^2 \phi)^2}$

Simplifying this expression, we obtain:


$|\vec{X}_\phi \times \vec{X}_\theta|(p) = \sqrt{a^2 b^2 c^2 \sin^2 \theta}$


Finally, we can compute the normal vector:


$\hat{N}(p) = \frac{\vec{X}_\phi \times \vec{X}_\theta}{|\vec{X}_\phi \times \vec{X}_\theta|(p)} = \left(\frac{bc \cos \phi \sin \theta}{\sqrt{a^2 b^2 c^2 \sin^2 \theta}}, \frac{-ac \sin \phi \cos \theta}{\sqrt{a^2 b^2 c^2 \sin^2 \theta}}, \frac{-ab \sin^2 \theta \cos \phi - ab \sin \theta \cos^2 \phi}{\sqrt{a^2 b^2 c^2 \sin^2 \theta}}\right)$

This expression represents the normal vector on the surface at point p.



```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the parameter values
a = 3
b = 2
c = 1

# Define the phi and theta values
phi = np.linspace(0, 2 * np.pi, 100)
theta = np.linspace(0, np.pi, 100)

# Create a meshgrid of phi and theta values
Phi, Theta = np.meshgrid(phi, theta)

# Compute the coordinates for the surface
X = a * np.sin(Theta) * np.cos(Phi)
Y = b * np.sin(Theta) * np.sin(Phi)
Z = c * np.cos(Theta)

# Define the coordinates for the point p
p = np.array([a * np.sin(np.pi/4) * np.cos(np.pi/3),b * np.sin(np.pi/4) * np.sin(np.pi/3),c * np.cos(np.pi/4)])

# Calculate the vectors X_phi, X_theta, and N_hat at point p
X_phi = np.array([-a * np.sin(np.pi/4) * np.sin(np.pi/3),b * np.sin(np.pi/4) * np.cos(np.pi/3),0])
X_theta = np.array([a * np.cos(np.pi/4) * np.cos(np.pi/3),b * np.cos(np.pi/4) * np.sin(np.pi/3),-c * np.sin(np.pi/4)])
N_hat = np.cross(X_phi, X_theta) / np.linalg.norm(np.cross(X_phi, X_theta))

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, alpha=0.8)

# Plot the point p and the vectors X_phi, X_theta, and N_hat
ax.quiver(p[0], p[1], p[2], X_phi[0], X_phi[1], X_phi[2], color='r', label='X_phi')
ax.quiver(p[0], p[1], p[2], X_theta[0], X_theta[1], X_theta[2], color='g', label='X_theta')
ax.quiver(p[0], p[1], p[2], N_hat[0], N_hat[1], N_hat[2], color='b', label='N*')
ax.scatter(p[0], p[1], p[2], color='k', s=50, label='Point p')

# Set the axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set the plot title
ax.set_title('Surface with Point p and Vectors')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

```python
import numpy as np

def ellipsoid_area(a, b, c, num_points):
    # Function to calculate the area of an ellipsoid using Monte Carlo method
    
    # Parameters:
    # a, b, c: Semi-axes of the ellipsoid
    # num_points: Number of random points to generate
    
    # Define the parametric equation of the ellipsoid
    def X(u, v):
        x = a * np.cos(u) * np.sin(v)
        y = b * np.sin(u) * np.sin(v)
        z = c * np.cos(v)
        return np.array([x, y, z])
    
    # Define the integration limits for u and v
    u_min, u_max = 0, 2 * np.pi
    v_min, v_max = 0, np.pi
    
    # Initialize variables
    area_sum = 0
    
    # Generate random points and estimate the area
    for _ in range(num_points):
        # Generate random values for u and v
        u = np.random.uniform(u_min, u_max)
        v = np.random.uniform(v_min, v_max)
        
        # Calculate the position vector X(u, v)
        position = X(u, v)
        
        # Calculate the magnitude of the cross product of the partial derivatives
        partial_u = np.array([-a * np.sin(u) * np.sin(v), b * np.cos(u) * np.sin(v), 0])
        partial_v = np.array([a * np.cos(u) * np.cos(v), b * np.sin(u) * np.cos(v), -c * np.sin(v)])
        cross_product = np.cross(partial_u, partial_v)
        magnitude = np.linalg.norm(cross_product)
        
        # Accumulate the magnitudes
        area_sum += magnitude
    
    # Calculate the average magnitude and estimate the area
    average_magnitude = area_sum / num_points
    integration_area = (u_max - u_min) * (v_max - v_min)
    estimated_area = average_magnitude * integration_area
    
    return estimated_area

# Example usage
a = 2  # Semi-axis along x-axis
b = 3  # Semi-axis along y-axis
c = 4  # Semi-axis along z-axis
num_points = 1000000  # Number of random points

estimated_area = ellipsoid_area(a, b, c, num_points)
print("Estimated area of the ellipsoid:", estimated_area)
```

The partial derivatives of $\vec{X}$ with respect to $\theta$ and t are:
